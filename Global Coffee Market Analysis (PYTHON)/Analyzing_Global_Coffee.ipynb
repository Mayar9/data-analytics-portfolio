# Task3

## task details 

* Our client is interested in looking at how Brazil's share of global coffee production has changed over time.
* Build a plot that global total production as well as Brazil's contribution to global production over time.

#### Way 1

```python
coffee_production.head()
coffee_T.head()

brazil_vs_restWorkd_ds = coffee_T.assign( rest_of_world = coffee_T.drop('Brazil', axis=1).sum(axis=1)).loc[:, ['Brazil', 'rest_of_world']].astype({'Brazil': 'float64', 'rest_of_world': 'float64'})
# brazil_vs_restWorkd_ds.dtypes

fig, ax = plt.subplots(figsize=(14,6))
ax.stackplot(brazil_vs_restWorkd_ds.index, 
             brazil_vs_restWorkd_ds['Brazil'], 
             brazil_vs_restWorkd_ds['rest_of_world'])

fig.legend(brazil_vs_restWorkd_ds.columns, loc='upper left')
fig.suptitle('Brazil share of Global Coffee Production 1990-2018', fontweight ='bold', fontsize=13)
ax.set_title('Brazil share increases over Growing Market', fontweight ='bold', fontsize=13)
ax.set_ylabel('Productions')
ax.xaxis.set_tick_params(rotation=30)
ax.spines[['top', 'right']].set_visible(False)
```

#### Way2

```python
Brazil_contribution =  coffee_production.query("total_production == 'Brazil'").reset_index(drop=True).melt(id_vars='total_production', value_name='Brazil', var_name='Year').set_index('Year').drop('total_production', axis=1)
Brazil_contribution


rest_of_world = pd.DataFrame({'Rest Of World': coffee_production.query("total_production != 'Brazil'").iloc[:, 1:].sum(axis=0)})
rest_of_world.index.name = 'Year'
rest_of_world


global_brazil_share = Brazil_contribution.merge(rest_of_world, how='inner', left_on='Year', right_on='Year')

fig, ax = plt.subplots(figsize=(14,6))
ax.stackplot(global_brazil_share.index, 
             global_brazil_share['Brazil'], 
             global_brazil_share['Rest Of World'])


fig.legend(global_brazil_share.columns, loc='upper left')
fig.suptitle('Brazil share of Global Coffee Production 1990-2018', fontweight ='bold', fontsize=13)
ax.set_title('Brazil share increases over Growing Market', fontweight ='bold', fontsize=13)
ax.set_ylabel('Productions')
ax.xaxis.set_tick_params(rotation=30)
ax.spines[['top', 'right']].set_visible(False)
```

#### Way3

```python
coffee_production.query("total_production == 'Brazil'").set_index('total_production')

rest_of_world_v2 = pd.DataFrame({'Rest Of World' :(coffee_production.query("total_production != 'Brazil'").iloc[:, 1:].sum(axis=0))}).rename_axis('year')
rest_of_world_v2.pivot_table(columns=rest_of_world_v2.index, values='Rest Of World')


dataset_global_share = (pd.concat([coffee_production.query("total_production == 'Brazil'").set_index('total_production'),
        rest_of_world_v2.pivot_table(columns=rest_of_world_v2.index, values='Rest Of World')])
        .rename_axis('Producers').reset_index()
        .melt(id_vars='Producers', var_name='Year', value_name='Overall Sales')
        .set_index('Year')
        .pivot_table(index='Year', columns= 'Producers', values='Overall Sales'))



fig, ax = plt.subplots(figsize=(14,6))
ax.stackplot(dataset_global_share.index, dataset_global_share['Brazil'], dataset_global_share['Rest Of World'], colors= ['blue', 'orange'])
fig.legend(dataset_global_share.columns, loc='upper left')
fig.suptitle('Brazil share of Global Coffee Production 1990-2018', fontweight ='bold', fontsize=13)
ax.set_title('Brazil share increases over Growing Market', fontweight ='bold', fontsize=13)
ax.set_ylabel('Productions')
ax.xaxis.set_tick_params(rotation=30)
ax.spines[['top', 'right']].set_visible(False)
```

### Task6

##### Coffee Consumption Vs. Prices

Compare coffee prices with coffee consumption in the following countries (prices countries). 

Is there a relationship between coffee consumption and the price these countries pay?

Plot a dual axis bar chart of price vs consumption totals for these countries.

```python
prices = pd.read_csv(r'E:\Python Data Visualization Matplotlib_Seaborn Masterclass\Project1\Global_Coffee_Market\Global_Coffee_Market\Datasets\retail-prices.csv', delimiter=',')
consumption = pd.read_csv(r'E:\Python Data Visualization Matplotlib_Seaborn Masterclass\Project1\Global_Coffee_Market\Global_Coffee_Market\Datasets\imports.csv', delimiter=',')
```

```python
consumption
# consumption.isna().sum()
```

```python
prices
# prices.isna().sum()
```

```python
prices['retail_prices'].isin(consumption['imports'])
```

```python
prices['retail_prices'].str.strip().isin(consumption['imports'].str.strip())
```

```python
prices['retail_prices'] = prices['retail_prices'].str.strip()
consumption['imports'] = consumption['imports'].str.strip()
```

```python
prices['retail_prices'].isin(consumption['imports'])
```

##### Way1

```python
prices_mean = prices.set_index('retail_prices').mean(axis=1)
prices_mean.index.name = 'country'
prices_mean.name = 'prices'
prices_mean

consumption_mean = consumption.set_index('imports').mean(axis=1)
consumption_mean.index.name = 'country'
consumption_mean.name = 'consumptions'
consumption_mean

prices_df = pd.DataFrame({'prices_mean': prices_mean})
consumption_df = pd.DataFrame({'consumption_mean': consumption_mean})

consump_prices_df = prices_df.merge(consumption_df, how='inner', left_on='country', right_on='country').sort_values('consumption_mean', ascending=False)
consump_prices_df

fig, ax = plt.subplots(figsize=(16,6))


width_ = 0.3
catagories = np.arange(len(consump_prices_df.index))
ax.bar(x= catagories-width_/2, height=consump_prices_df['consumption_mean'], width=width_, color='blue', alpha=0.4)
ax.set_ylabel("Consumption (k 60kg bags)", fontsize=12)

ax2 = ax.twinx()
ax2.bar(x= catagories+width_/2, height=consump_prices_df['prices_mean'], width=width_, color='orange', alpha=0.4)
ax2.set_ylabel("Average Price Paid Per Bag (USD)", fontsize=12)


ax.set_xticks(catagories, consump_prices_df.index)
ax.xaxis.set_tick_params(rotation=30)

fig.gca().spines[['top', 'right']].set_visible(False)
fig.legend(['Consumption', 'Prices'], bbox_to_anchor=(0.9, 0.88))
fig.suptitle('Prices Paid Vs Consumptions of Importing Nations', fontsize=18, fontweight='bold')

```

##### Way2

```python
prices_consumption_df  = pd.DataFrame(prices_mean).merge(consumption_mean, how='inner', left_on='country', right_on='country').sort_values('consumptions', ascending=False)
prices_consumption_df


fig, ax = plt.subplots(figsize=(16,6))


width_ = 0.3
catagories = np.arange(len(prices_consumption_df.index))
ax.bar(x= catagories-width_/2, height=prices_consumption_df['consumptions'], width=width_, color='blue', alpha=0.4)
ax.set_ylabel("Consumption (k 60kg bags)", fontsize=12)

ax2 = ax.twinx()
ax2.bar(x= catagories+width_/2, height=prices_consumption_df['prices'], width=width_, color='orange', alpha=0.4)
ax2.set_ylabel("Average Price Paid Per Bag (USD)", fontsize=12)


ax.set_xticks(catagories, prices_consumption_df.index)
ax.xaxis.set_tick_params(rotation=30)

fig.gca().spines[['top', 'right']].set_visible(False)
fig.legend(['Consumption', 'Prices'], bbox_to_anchor=(0.9, 0.88))
fig.suptitle('Prices Paid Vs Consumptions of Importing Nations', fontsize=18, fontweight='bold')
```
